```python
from sklearn.linear_model import LinearRegression
import pandas as pd
df = pd.read_csv('2023-10-18 10_09pm.csv')
df
```

```python

df.drop(['NORMALIZEDPREVIOUSMETERNUMBER','PODNUMBER','NEXTREADDATE','ACCOUNTNUMBER','METERNUMBER'], axis=1,inplace=True)
df
```

```python
count_nan_in_df = df.isnull().sum()
print (count_nan_in_df)
```

```python
df = df.dropna(subset=['METERREADDATE'], how='any')
```

```python
df
```

```python
df['PREVREADDATE'] = pd.to_datetime(df['PREVREADDATE'])
```

```python
df['METERREADDATE'] = pd.to_datetime(df['METERREADDATE'])
df
```

```python
print(df.dtypes)
```

```python
df['month_year'] = df['METERREADDATE'].dt.strftime('%m-%Y')
```

```python
df
```

```python
df['NORMALIZEDACCOUNTNUMBER'].nunique()
```

```python
df['INTERVAL'] = (df['METERREADDATE'] - df['PREVREADDATE']).dt.days
df
```

```python
value_counts = df['NORMALIZEDACCOUNTNUMBER'].value_counts()
value_counts
```

```python
acc_numbers_to_drop = value_counts[value_counts < 6].index
acc_numbers_to_drop
```

```python
df = df[~df['NORMALIZEDACCOUNTNUMBER'].isin(acc_numbers_to_drop)]
df
```

```python
new_month_wise_df2=df.pivot_table(values='INTERVAL', index=['month_year'], columns=['NORMALIZEDACCOUNTNUMBER','CUSTOMERALIAS','PROVIDERALIAS'],  aggfunc=sum).T.reset_index()
new_month_wise_df2
```

```python
column_headers = list(new_month_wise_df2.columns)
column_headers1=[]
for i in range(len(column_headers)):
    if '-' in column_headers[i]:
        column_headers1.append(column_headers[i])
column_headers1
```

```python
from datetime import datetime
column_headers1.sort(key=lambda date: datetime.strptime(date, "%m-%Y"))
column_headers1
```

```python
column_headers1.append('NORMALIZEDACCOUNTNUMBER')
```

```python
column_headers1
```

```python
column_headers1.sort(key=lambda date: datetime.strptime(date, "%m-%Y"))
```

```python
new_month_wise_df1 =new_month_wise_df2[column_headers1].apply(lambda row: row.fillna(row.median()), axis=1)
```

```python
new_month_wise_df1
```

```python
df_merged = new_month_wise_df1.merge(new_month_wise_df2[['NORMALIZEDACCOUNTNUMBER','PROVIDERALIAS']])
```

```python
df_merged
```

```python
prepared_Set = df_merged
prepared_Set = prepared_Set.drop(['10-2022'],axis=1)
prepared_Set
```

```python
from sklearn.preprocessing import OneHotEncoder

cat_cols = ["PROVIDERALIAS"]
cat_cols_encoded = []
for col in cat_cols:
  cat_cols_encoded += [f"{col[0]}_{cat}" for cat in list(prepared_Set[col].unique())]

cat_cols_encoded
```

```python
oh_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_cols = oh_encoder.fit_transform(prepared_Set[cat_cols])
df_enc = pd.DataFrame(encoded_cols, columns=cat_cols_encoded)
df_oh = prepared_Set.join(df_enc)
df_oh
```

```python
prepared_Set = df_oh.drop(['NORMALIZEDACCOUNTNUMBER','PROVIDERALIAS'],axis=1)
prepared_Set
```

```python
data_x=prepared_Set.loc[:, prepared_Set.columns != '10-2023']
data_y=prepared_Set['10-2023']
```

```python
data_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test = train_test_split(data_x,data_y,test_size =0.3)
# print the data
x_train
```

```python
clf = LinearRegression()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
clf.score(x_test,y_test)
```

```python
import numpy as np
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
pred_train_lr= clf.predict(x_train)
print(np.sqrt(mean_squared_error(y_train,pred_train_lr)))
print(r2_score(y_train, pred_train_lr))

pred_test_lr= clf.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,pred_test_lr)))
print(r2_score(y_test, pred_test_lr))
```
